```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and a set of subplots
fig, axs = plt.subplots(1, 5, figsize=(20, 5))

# Set common properties for all axes
for ax in axs:
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.axis('off')

# Memory display
axs[0].add_patch(patches.Circle((3, 5), 2, edgecolor='gray', facecolor='red', linestyle='--'))
axs[0].add_patch(patches.Circle((7, 5), 2, edgecolor='gray', facecolor='blue', linestyle='--'))
axs[0].text(3, 3, 'Freq. A', ha='center')
axs[0].text(7, 3, 'Freq. B', ha='center')
axs[0].text(5, 1, '1000ms', ha='center')

# Delay period
axs[1].add_patch(patches.Circle((3, 5), 2, edgecolor='gray', linestyle='--'))
axs[1].add_patch(patches.Circle((7, 5), 2, edgecolor='gray', linestyle='--'))
axs[1].text(5, 1, '1500-1900ms', ha='center')

# Retro-Cue
axs[2].add_patch(patches.Circle((3, 5), 2, edgecolor='gray', linestyle='--'))
axs[2].add_patch(patches.Circle((7, 5), 2, edgecolor='gray', linestyle='--'))
axs[2].add_patch(patches.Circle((3, 5), 1.5, edgecolor='red', facecolor='red', alpha=0.5))
axs[2].text(5, 1, '150ms', ha='center')

# Delay period
axs[3].add_patch(patches.Circle((3, 5), 2, edgecolor='gray', linestyle='--'))
axs[3].add_patch(patches.Circle((7, 5), 2, edgecolor='gray', linestyle='--'))
axs[3].text(5, 1, '1500ms', ha='center')

# Probe
axs[4].add_patch(patches.Circle((3, 5), 2, edgecolor='gray', linestyle='--'))
axs[4].add_patch(patches.Circle((7, 5), 2, edgecolor='gray', linestyle='--'))
axs[4].add_patch(patches.Rectangle((4.5, 4.5), 1, 1, angle=45, edgecolor='black', facecolor='none'))
axs[4].text(5, 1, 'Probe: CWise/\nCCWise', ha='center')

# Show the plot
plt.show()
```
